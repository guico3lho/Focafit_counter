```python
import re
from datetime import datetime, timedelta

def extract_messages_by_date_range(file_path, start_date, end_date):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.readlines()

    messages = []
    for line in text:
        match = re.search(r'(\d{2}/\d{2}/\d{4}),', line)
        if match:
            message_date = datetime.strptime(match.group(1), '%d/%m/%Y')
            if start_date <= message_date <= end_date:
                messages.append(line)

    return messages

```

```python
file_path = 'path/to/your/exported/chat.txt'
today = datetime.strptime('15/05/2023', '%d/%m/%Y')
start_date = today - timedelta(days=7)
end_date = today - timedelta(days=1)

messages = extract_messages_by_date_range(file_path, start_date, end_date)

for message in messages:
    print(message)
```
